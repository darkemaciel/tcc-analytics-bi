```python
# Nova string de conexão
conn_str = f'DRIVER=SQL Server;SERVER={server_name};DATABASE={database};TRUSTED_CONNECTION={trusted_connection}'
conn = pyodbc.connect(conn_str)
```

```python
# Criar um cursor para executar comandos SQL
cursor = conn.cursor()
```

```python
# Supondo que você tenha uma conexão com o SQL Server
engine = create_engine(f'mssql+pyodbc:///?odbc_connect={conn_str}')

# Lendo o arquivo Parquet de Receitas
df = pd.read_parquet(r'C:\Users\Darke\Documents\Especialização - Dados\Analitycs e BI\TCC\arquivos\despesas\despesas_unificadas.parquet')

# Defina o tamanho do lote (chunk size)
chunk_size = 5000
total_rows = len(df)

# Crie uma barra de progresso
progress_bar = tqdm(total=total_rows, desc='Ingestão para SQL')

# Itere sobre o DataFrame em partes (chunks)
for i in range(0, total_rows, chunk_size):
    df_part = df.iloc[i:i + chunk_size]
    
    # Inserir parte do DataFrame no banco de dados
    df_part.to_sql('tb_stg_despesas', con=engine, index=False, if_exists='append')

    # Atualizar a barra de progresso
    progress_bar.update(len(df_part))

# Fechar a barra de progresso quando a ingestão estiver concluída
progress_bar.close()
```

```python
# Fechar a conexão
cursor.close()
conn.close()
```
